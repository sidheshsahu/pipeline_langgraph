```python
import os
from langgraph.graph import StateGraph, MessagesState, START, END
from typing import TypedDict, Annotated, List
```

```python
class MathState(TypedDict):
    num_1: int
    num_2: int
    addition: int
    subtract: int
    multiplication: int
```

```python
def add_node(state: MathState):
    addition = state["num_1"] + state["num_2"]
    return {"addition": addition}
```

```python
def subtract_node(state: MathState):
    subtract = state["num_1"] - state["num_2"]
    return {"subtract": subtract}
```

```python
def multiplication_node(state: MathState):
    multiplication = state["num_1"] * state["num_2"]
    return {"multiplication": multiplication}
```

```python
graph = StateGraph(MathState)
graph.add_node("add", add_node)
graph.add_node("subtract", subtract_node)
graph.add_node("multiplication", multiplication_node)
graph.add_edge(START, "add")
graph.add_edge("add", "subtract")
graph.add_edge("subtract", "multiplication")
graph.add_edge("multiplication", END)
app = graph.compile()
```

```python
output = app.invoke({"num_1": 10, "num_2": 5})
print(output)
```
